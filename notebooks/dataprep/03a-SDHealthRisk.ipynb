```python
import os
import pandas as pd
from pathlib import Path
from py2neo import Graph
```

```python
pd.options.display.max_rows = None  # display all rows
pd.options.display.max_columns = None  # display all columsns
```

```python
NEO4J_HOME = Path(os.getenv('NEO4J_HOME'))
print(NEO4J_HOME)
```

### Load San Diego County Health Risk data

```python
df = pd.read_csv("../../reference_data/risks_by_tracts_san_doegp.csv", dtype='str')
df.fillna('', inplace=True)
```

```python
df.rename(columns={'TractFIPS':'tract',
                   'High Blood Pressure %': 'highBloodPressurePct',
                   'Cancer %':'cancerPct',
                   'Asthma %':'asthmaPct',
                   'Hearth disease %':'heartDiseasePct',
                   'Lung disease( COPD) %':'lungDiseaseCOPDPct',
                   'Diabetes %':'diabetesPct',
                   'High Cholesterol %':'highCholesterolPct',
                   'Kidney disease %':'kidneyDiseasePct',
                   'Mental health %':'mentalHealthPct',
                   'Obesity %':'obesityPct',
                   'Poor physical health %':'poorPhysicalHealthPct',
                   'Stroke %':'strokePct'
                  }, 
          inplace=True)
```

#### Prepend '0' to match US Census Tract convention

```python
df['tract'] = '0' + df['tract']
```

```python
df.head(5)
```

### Save selected data

```python
df = df[['tract', 'highBloodPressurePct', 'cancerPct', 'asthmaPct', 'heartDiseasePct', 'lungDiseaseCOPDPct', 
         'diabetesPct', 'highCholesterolPct', 'kidneyDiseasePct', 'mentalHealthPct', 'obesityPct', 
         'poorPhysicalHealthPct', 'strokePct']]
df.to_csv(NEO4J_HOME / 'import/03a-SDHealthRisk.csv', index = False)
```
